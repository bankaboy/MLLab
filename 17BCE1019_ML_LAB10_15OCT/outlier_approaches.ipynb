<h1>Standard Deviation</h1>
If any data point is more than 3 times the standard deviation, then those points are very likely to be anomalous or outliers. This is applicable for linear data

```python
import numpy as np
import matplotlib.pyplot as plt
import random
random.seed(1)
import scipy 
anomalies = []

# multiply and add by random numbers to get some real values
data = np.random.randn(50000)  * 20 + 20

# Function to Detection Outlier on one-dimentional datasets.
def find_anomalies(random_data):
    # Set upper and lower limit to 3 standard deviation
    random_data_std = scipy.std(random_data)
    random_data_mean = scipy.mean(random_data)
    anomaly_cut_off = random_data_std * 3
    
    lower_limit  = random_data_mean - anomaly_cut_off 
    upper_limit = random_data_mean + anomaly_cut_off
    print(lower_limit)
    # Generate outliers
    for outlier in random_data:
        if outlier > upper_limit or outlier < lower_limit:
            anomalies.append(outlier)
    return anomalies

find_anomalies(data)
```

<h1>Boxplots</h1>
These are graphical representation of numerical data through quantiles. Any data points that show above or below the whiskers, can be considered outliers or anomalous.

```python
import seaborn as sns
import matplotlib.pyplot as plt

random_data = np.random.randn(50000)  * 20 + 20
sns.boxplot(data=random_data)
```

<h1>DBScan Clustering</h1>
DBScan is a clustering algorithm that’s used cluster data into groups. It is also used as a density-based anomaly detection method with either single or multi-dimensional data. 

```python
from sklearn.cluster import DBSCAN
random.seed(1)
random_data = np.random.randn(50000,2)  * 20 + 20

outlier_detection = DBSCAN(min_samples = 2, eps = 3)
clusters = outlier_detection.fit_predict(random_data)
list(clusters).count(-1)
```

<h1>Isolation Forest</h1>
It explicitly isolates anomalies instead of profiling and constructing normal points and regions by assigning a score to each data point. It takes advantage of the fact that anomalies are the minority data points and that they have attribute-values that are very different from those of normal instances.

```python
from sklearn.ensemble import IsolationForest
import numpy as np
np.random.seed(1)
random_data = np.random.randn(50000,2)  * 20 + 20

clf = IsolationForest( behaviour = 'new', max_samples=100, random_state = 1, contamination= 'auto')
preds = clf.fit_predict(random_data)
for i in range(len(preds)):
    print("Outlier prediction for element ",i," is ",preds[i])
```
