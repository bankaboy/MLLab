<h3> Perfom Missing Value Analysis</h3>
Missing value analysis helps address several concerns caused by incomplete data. If cases with missing values are systematically different from cases without missing values, the results can be misleading. Also, missing data may reduce the precision of calculated statistics because there is less information than originally planned. Another concern is that the assumptions behind many statistical procedures are based on complete cases, and missing values can complicate the theory required.

```python
import pandas as pd
import numpy as np
mydata = pd.read_csv("UCI_Credit_Card.csv")
mydata
```

```python
from sklearn.preprocessing import Imputer
values = mydata.values
imputer = Imputer()
transformed_data = imputer.fit_transform(values)
transformed_data
```

<h3>Normalize Data</h3>
Data normalization gets rid of a number of anomalies that can make analysis of the data more complicated. Some of those anomalies can crop up from deleting data, inserting more information, or updating existing information. Once those errors are worked out and removed from the system, further benefits can be gained through other uses of the data and data analytics.

```python
X = mydata[mydata.columns[:24]]
y = mydata.default_payment_next_month

from sklearn import preprocessing
from sklearn.model_selection import *

x_train, x_test, y_train, y_test = train_test_split(X, y , train_size = 0.7, random_state =  90)

#Select numerical columns which needs to be normalized
train_norm = x_train[x_train.columns[0:24]]
test_norm = x_test[x_test.columns[0:24]]

# Normalize Training Data 
std_scale = preprocessing.StandardScaler().fit(train_norm)
x_train_norm = std_scale.transform(train_norm)

x_train_norm
```

<h3>Check for outliers</h3>
Isolation Forest:<br>
It explicitly isolates anomalies instead of profiling and constructing normal points and regions by assigning a score to each data point. It takes advantage of the fact that anomalies are the minority data points and that they have attribute-values that are very different from those of normal instances.

```python
from sklearn.ensemble import IsolationForest

def column(matrix, i):
    return [row[i] for row in matrix]

clf = IsolationForest( behaviour = 'new', max_samples=100, random_state = 1, contamination= 'auto')
preds = clf.fit_predict(x_train_norm)
for j in range(len(preds)):
    if preds[j] == -1:
        print("Outiler at element ",j)
```

<h3>Principal Component Analysis</h3>
Principal component analysis (PCA) is a statistical procedure that uses an orthogonal transformation to convert a set of observations of possibly correlated variables (entities each of which takes on various numerical values) into a set of values of linearly uncorrelated variables called principal components.

```python
# performing preprocessing part 
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler() 
  
x_train = sc.fit_transform(x_train) 
x_test = sc.transform(x_test)

# Applying PCA function on training 
# and testing set of X component 
from sklearn.decomposition import PCA 

pca = PCA(n_components = 2) 

x_train = pca.fit_transform(x_train) 
x_test = pca.transform(x_test) 

explained_variance = pca.explained_variance_ratio_ 
explained_variance
```
